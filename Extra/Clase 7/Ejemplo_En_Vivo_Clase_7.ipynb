# Conjunto

```python
conj1 = {1,2,4,"Nico"}
conj2 = {2,6,"Pepe"}
conj3 = {1,2,4,"Nico",8,11}
```

```python
#Esta función comprueba si el set es distinto a otro set, es decir, si no hay ningún ítem en común entre ellos
conj1.isdisjoint(conj2)
```

```python
#Esta función comprueba si el set es subset de otro set, es decir, si todos sus ítems están en el otro conjunto
conj1.issubset(conj3)
```

```python
#Esta función es muy similar al issubset, la diferencia es que esta comprueba si el set es contenedor de otro set, es decir, si contiene todos los ítems de otro set. 
conj3.issuperset(conj1)
```

```python
#Esta función une un set con otro, y devuelve el resultado en un nuevo set. 
conj1.union(conj2)
```

```python
#Esta función encuentra todos los elementos no comunes entre dos set, es decir, nos devuelve un set de ítems diferentes entre cada set. 
conj3.difference(conj1)
```

```python
conj1.intersection(conj2)
```

# Diccionarios

```python
diccionario = {"Nombre":"Nicolas", "Edad":32, "Domicilio":"Av.Pellegrini 564"}
```

```python
diccionario.keys()
```

```python
diccionario.values()
```

```python
diccionario.items()
```

Cuando iteramos en diccionarios podemos hacerlo a través de las claves, valores o ítems (pares clave-valor). Para ello podemos apoyarnos por ejemplo del uso de un bucle for.

```python
diccionario_2 = {"Nombre":"Nicolas", "Edad":32, "Domicilio":"Av.Pellegrini 564", 'Telefono': 99999999, 'Nacionalidad': 'Argentino'}

# Iterar en llaves
for key in diccionario_2.keys():
    print(key)

# Iterar en valores
for value in diccionario_2.values():
    print(value)

# Iterar en ítems
for key, value in diccionario_2.items():
    print(f'Clave {key}, valor {value}')
```

```python
#Tambien podemos apoyarnos en el metodo get!
print(diccionario_2.get("Nombre")) 
```

```python
print(diccionario_2.get("Nacionalidad"))
```

```python
print(diccionario_2.get("Edad"))
```

```python
print(diccionario_2.get("EstadoCivil", "Ese elemento no existe dentro de nuestro diccionario"))
```
