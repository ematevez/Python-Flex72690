# Repaso 2: Dada esta lista de números encontrar el valor mínimo y la ubicación.   [21,33,-25,85,-3.6, 27]

```python
# -25

lista = [21,33,-25,85,-3.6, 27]

minimo = lista[0]

for x in lista:

  if ( x < minimo):

    minimo = x


print(f"MINIMO: {minimo}")

#Tarea para la casa !!!!!!!! ---- en qué posición estaba el -25
```
