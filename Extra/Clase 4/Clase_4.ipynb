# Control de flujo de decisión 

# if

```python
#Ejemplo 1
x = 10

if x > 4:
  print("¡La condición es verdadera!")
```

```python
#Ejemplo 2
numero = 6

if (  numero%2 == 0 ):
  print("Tu numero  es par :)")

if ( numero%3 == 0):
  print("Tu numero es divisible por 3 :)")

if (numero > 10):
  print("Tu numero es muy grande :( ")
```

```python
#Ejemplo 3
a = 10
if a > 5 and a < 15:
    print("Mayor que 5 y menos que 15")
```

# Else

```python
#Ejemplo 1
x = 10
if x == 10:
    print("Es 10")
else:
    print("No es 10")
```

```python
#Ejemplo 2
a = 5
b = 10

if a < b:
    print(" b es mayor que a")
else:
    print("a es mayor que b")
```

# Elif

```python
#Ejemplo 1
x = 5
if x == 5:
    print("Es 5")
elif x == 6:
    print("Es 6")
elif x == 7:
    print("Es 7")
```

```python
#Ejemplo 2
x = 5
if x == 5:
    print("Es 5")
elif x == 6:
    print("Es 6")
elif x == 7:
    print("Es 7")
else:
    print("Es otro")
```

```python
#Mas ejemplos :)
comando =  "SALUDO"

if (comando == "ENTRAR"):

    print("Bienvenido al sistema.")
  
elif (comando == "SALUDO"):
  
    print("Hola! ¿Cómo estás?")
  
elif (comando == "SALIR"):
    print("Saliendo del sistema.")
  
else:
    print("No se reconoce el comando.")
```
