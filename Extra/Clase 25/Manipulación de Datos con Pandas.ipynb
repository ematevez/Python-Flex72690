## Librerías en Python

¿Qué son las Librerías? Las librerías, son paquetes con funciones que ya están programadas por alguien más y realmente nos van a resultar de mucha utilidad a la hora de programar. Existen muchísimas librerías en Python. 

Algunas de las librerías más importantes en Python son:

* **NumPy**: Proporciona algunas funcionalidades matemáticas avanzadas para Python. 
* **Pandas**: Permite realizar tareas de manipulación de datos sobre DataFrames.
* **Matplotlib**: Matplotlib es la biblioteca más utilizada para crear visualizaciones de datos simples pero potentes.
* **Plotly**: Plotly es una maravillosa biblioteca de visualización de datos interactiva basada en la web. 
* **Seaborn**: Seaborn es también una biblioteca de visualización muy útil, basada en Matplotlib, etc.

Por supuesto que existen muchísimas librerías más, como por ejemplo SciKit-Learn y TensorFlow para trabajar en Machine Learning e Inteligencia Artificial respectivamente. 

¿Cómo trabajamos con las Librerías? lo primero que tenemos que hacer es instalarlas (generalmente a través de la terminal de consola). Luego, utilizaremos el comando **import**.
