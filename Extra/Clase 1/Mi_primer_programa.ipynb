Trabajas en Coderhouse y te piden crear un programa que calcule la nota final de estudiantes del curso de Python. La nota final se calcula en base a tres notas previas de las cuales, cada una corresponde un porcentaje distinto de la nota final. Los porcentajes se detallan a continuación:

nota_1 cuenta como el 20% de la nota final

nota_2 cuenta con el 30% de la nota final

nota_3 cuenta con el 50% de la nota final

Importante: el programa deberá solicitar al usuario que se ingrese el valor de la nota mediante la funcion input()

```python
#Solicitud de datos al usuario
```

```python
nota_1 = int(input("Ingresa la nota 1: "))
nota_1
```

```python
nota_2 = int(input("Ingresa la nota 2: "))
nota_2
```

```python
nota_3 = int(input("Ingresa la nota 3: "))
nota_3 
```

```python
#Asignacion de porcentajes
nota_1 = nota_1 * 0.20
nota_2 = nota_2 * 0.30
nota_3 = nota_3 * 0.50
```

```python
nota_total = nota_1 + nota_2 + nota_3
nota_total
```
